```python
import fastf1
import os
import psycopg2

fastf1.Cache.enable_cache("/Users/noahhowren/Documents/pitlane/database/fastf1cache/")

conn = psycopg2.connect(database="F1_DB",
                        user = "postgres", password = "postgres",
                        host="127.0.0.1", port="5432")
conn.autocommit = True
cursor = conn.cursor()

sql = '''SELECT * FROM RACES
        WHERE YEAR = 2022'''

cursor.execute(sql)

races = cursor.fetchall()

print(races)


for row in races:
    fp1= fastf1.get_session(2022, row[4], 'FP1')
    fp1.load()
    fp2 = fastf1.get_session(2022, row[4], 'FP2')
    fp2.load()
    fp3 = fastf1.get_session(2022, row[4], 'FP3')
    fp3.load()
    q = fastf1.get_session(2022, row[4], 'Q')
    q.load()
    #fastf1.get_session(2022, row[4], 'S')
    r = fastf1.get_session(2022, row[4], 'R')
    r.load()
```
